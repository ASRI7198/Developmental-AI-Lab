```python
# @title Initialize the turtle environment

BORDER_WIDTH = 20

class ColabTurtleEnvironment:

    def __init__(self):
        """ Creating the Turtle window """
        bgcolor("lightGray")
        penup()
        goto(window_width() / 2, window_height()/2)
        face(0)
        pendown()
        color("green")

    def outcome(self, action):
        """ Enacting an action and returning the outcome """
        _outcome = 0
        for i in range(10):
            # _outcome = 0
            if action == 0:
                # move forward
                forward(10)
            elif action == 1:
                # rotate left
                left(4)
                forward(2)
            elif action == 2:
                # rotate right
                right(4)
                forward(2)

            # Bump on screen edge and return outcome 1
            if xcor() < BORDER_WIDTH:
                goto(BORDER_WIDTH, ycor())
                _outcome = 1
            if xcor() > window_width() - BORDER_WIDTH:
                goto(window_width() - BORDER_WIDTH, ycor())
                _outcome = 1
            if ycor() < BORDER_WIDTH:
                goto(xcor(), BORDER_WIDTH)
                _outcome = 1
            if ycor() > window_height() - BORDER_WIDTH:
                goto(xcor(), window_height() -BORDER_WIDTH)
                _outcome = 1

            # Change color
            if _outcome == 0:
                color("green")
            else:
                # Finit l'interaction
                color("red")
                # if action == 0:
                #     break
                if action == 1:
                    for j in range(10):
                        left(4)
                elif action == 2:
                    for j in range(10):
                        right(4)
                break

        return _outcome
```

```python
#----------------------------------Agent 2--------------------------------------#
import random


class Agent2:
    def __init__(self, valence, Max_Ennui):
        """ Creating our agent """
        self.valence = valence
        self.action = 0
        self.pred_outcome = 0

        self.Max_Ennui = Max_Ennui
        self.memoire = {}
        self.listePrefere = [0, 0]  # Action et Outcome
        self.Numbre_action_Prefere = 0
        self.actionfinal = 0

    def Action(self, outcome):
        """ Computing the next action to enact """
        a = self.valence[self.listePrefere[0]][self.listePrefere[1]]
        b = self.valence[self.action][outcome]
        if b > a:
            self.listePrefere[0] = self.action
            self.listePrefere[1] = outcome
            self.Numbre_action_Prefere = 0

        self.action = self.listePrefere[0]

        if self.Numbre_action_Prefere >= self.Max_Ennui:
            self.Numbre_action_Prefere = 0
            self.action = 1 - self.action

        self.Update_Memoire(outcome)
        if self.listePrefere[0] == self.action:
            self.Numbre_action_Prefere += 1

        self.actionfinal = self.action

        return self.action

    def Update_Memoire(self, outcome):
        if self.action not in self.memoire.keys():
            self.memoire[self.action] = outcome
        elif self.pred_outcome != outcome:
            self.memoire[self.actionfinal] = outcome

    def Message(self, outcome):
        print(f"Action: {self.action}, Prediction: {self.pred_outcome}, Outcome: {outcome}, "
              f"Prediction: {self.pred_outcome == outcome}, Valence: {self.valence[self.action][outcome]}")


class Environment1:
    """ In Environment 1, action 0 yields outcome 0, action 1 yields outcome 1 """

    def outcome(self, _action):
        # return int(input("entre 0 1 ou 2"))
        if _action == 0:
            return 0
        else:
            return 1


class Environment2:
    """ In Environment 1, action 0 yields outcome 0, action 1 yields outcome 1 """

    def outcome(self, _action):
        # return int(input("entre 0 1 ou 2"))
        if _action == 0:
            return 1
        else:
            return 0


valence1 = [[-1, 1], [-1, 1]]
valence2 = [[1, -1], [1, -1]]
agent = Agent2(valence2, 4)
env = Environment1()
outcome = 0
```

```python
#----------------------------------Agent 3--------------------------------------#
import random


class Agent3:
    def __init__(self, valence):
        """ Creating our agent """
        self.valence = valence
        self.action = 0
        self.pred_outcome = 0

        self.Max_Ennui = 1
        self.memoire = {}
        self.listePrefere = [0, 0]  # Action et Outcome
        self.Numbre_action_Prefere = 0
        self.actionfinal = 0

    def Action(self, outcome):
        """ Computing the next action to enact """
        a = self.valence[self.listePrefere[0]][self.listePrefere[1]]
        b = self.valence[self.action][outcome]
        if b > a:
            self.listePrefere[0] = self.action
            self.listePrefere[1] = outcome
            self.Numbre_action_Prefere = 0

        self.action = self.listePrefere[0]

        if self.Numbre_action_Prefere >= self.Max_Ennui:
            self.Numbre_action_Prefere = 0
            if self.action==0:
                self.action = random.choice([1,2])
            elif self.action==1:
                self.action = random.choice([0,2])
            elif self.action==2:
                self.action = random.choice([0,1])

        self.Update_Memoire(outcome)
        if self.listePrefere[0] == self.action:
            self.Numbre_action_Prefere += 1

        self.actionfinal = self.action

        return self.action

    def Update_Memoire(self, outcome):
        if self.action not in self.memoire.keys():
            self.memoire[self.action] = outcome
        elif self.pred_outcome != outcome:
            self.memoire[self.actionfinal] = outcome

    def Message(self, outcome):
        
        print("[",self.action,"][",outcome,"]")
        print(f"Action: {self.action}, Prediction: {self.pred_outcome}, Outcome: {outcome}, "
              f"Prediction: {self.pred_outcome == outcome}, Valence: {self.valence[self.action][outcome]}")
```
